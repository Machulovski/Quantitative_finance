```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m,rev_stdev,iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev*np.random.normal(0.6,0.1))
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
print(COGS.mean())
print(COGS.std())
```

```python
Gross_profit = rev + COGS
plt.figure(figsize=(15,6))
plt.plot(Gross_profit)
plt.show()
print("GP mean: "+str(Gross_profit.mean()))
print("GP std dev: "+str(Gross_profit.std()))
```

```python
plt.figure(figsize=(10,6))
plt.hist(Gross_profit, bins = 50)
plt.show()
```
